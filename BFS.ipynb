```python
from collections import deque

def bfs(graph, start):
    visited = []
    queue = deque()

    queue.append(start)

    while queue:
        node = queue.popleft()

        if node not in visited:
            visited.append(node)

            for neighbour in graph[node]:
                queue.append(neighbour)

    return visited
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

print("BFS Traversal:", bfs(graph, 'A'))
```
